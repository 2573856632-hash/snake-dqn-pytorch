```python
# simple_test.py
import numpy as np
import torch
import matplotlib.pyplot as plt

# 
def test_basic_functionality():
    print("Testing basic PyTorch functionality...")
    
    # PyTorch
    x = torch.randn(3, 3)
    print(f"PyTorch test tensor: {x}")
    print(f"PyTorch version: {torch.__version__}")
    
    # numpy
    arr = np.random.randn(3, 3)
    print(f"NumPy test array shape: {arr.shape}")
    
    # matplotlib
    plt.figure(figsize=(8, 4))
    
    # 
    episodes = 100
    scores = [np.random.randint(0, 100) for _ in range(episodes)]
    
    # 
    window_size = 10
    smoothed_scores = []
    for i in range(len(scores)):
        start = max(0, i - window_size // 2)
        end = min(len(scores), i + window_size // 2 + 1)
        smoothed_scores.append(np.mean(scores[start:end]))
    
    plt.subplot(1, 2, 1)
    plt.plot(scores, alpha=0.3, label='Raw')
    plt.plot(smoothed_scores, label='Smoothed')
    plt.title('Simulated Training Progress')
    plt.xlabel('Episode')
    plt.ylabel('Score')
    plt.legend()
    plt.grid(True)
    
    plt.subplot(1, 2, 2)
    plt.hist(scores, bins=20, alpha=0.7)
    plt.title('Score Distribution')
    plt.xlabel('Score')
    plt.ylabel('Frequency')
    plt.grid(True)
    
    plt.tight_layout()
    plt.savefig('test_results.png', dpi=150, bbox_inches='tight')
    plt.show()
    
    print("Basic functionality test completed!")
    print("Check 'test_results.png' for the generated plot")

if __name__ == "__main__":
    test_basic_functionality()
```
